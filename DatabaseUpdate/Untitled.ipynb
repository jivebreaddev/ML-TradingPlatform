```python
params = urllib.parse.quote("DRIVER={ODBC Driver 13 for SQL Server};SERVER=localhost;DATABASE=StocksDMW2;Trusted_Connection=yes")
engine1 = sqlalchemy.create_engine('mssql+pyodbc:///?odbc_connect=%s' % params)
```

```python
symbols = """
select distinct(name) from stockdatadaily order by name
"""
symbol_list = pd.read_sql(symbols, engine1)['name'].tolist()
```

```python
pbar = tqdm(total = len(symbol_list))

for i in symbol_list:
    print(i)
    symbol_data = """
    select * from stockdatadaily where name = '%s'
    """ % i
    data = pd.read_sql(symbol_data, engine1)
    data.reset_index()
    
    dataFrame = pd.DataFrame()
    dataFrame['date'] = data['date']
    dataFrame['1. open'] = data['open']
    dataFrame['2. high'] = data['high']
    dataFrame['3. low'] = data['low']
    dataFrame['4. close'] = data['close']
    dataFrame['5. volume'] = data['volume']
    dataFrame['Name'] =  data['Name']
    
    symbol_data_2 = """
    select * from stockdatadaily where name = '%s'
    """ % i
    data2 = pd.read_sql(symbol_data, engine2)
    
    c = pd.concat([dataFrame, data2 , data2]).drop_duplicates(keep=False)
    
    
    c.to_sql('StockDataDaily', con=engine2, if_exists='replace', index=False)
    
    pbar.update(1)
pbar.close()
```
